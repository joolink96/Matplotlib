# 2.축

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic' #Windows
matplotlib.rcParams['font.size']=15 #글자크기
matplotlib.rcParams['axes.unicode_minus']=False #한글 폰트를 사용시, 마이너스 글자 깨지지 않도록함
```

```python
x=[1,2,3]
y=[2,4,8]
```

```python
plt.plot(x,y)
plt.title('꺽은선 그래프',fontdict={'family':'HYGungSo-Bold','size':20}) #개별 폰트 설정
```

```python
plt.plot(x,y)
plt.xlabel('X축',color='red')
plt.ylabel('Y축',color='#00aa00') # RGB중 R,B는 없음 Green만 존재
```

```python
plt.plot(x,y)
plt.xlabel('X축',color='red',loc='right') #left,center,right
plt.ylabel('Y축',color='#00aa00',loc='top') # top,center,bottom 
```

```python
plt.plot(x,y)
plt.xticks([1,2,3]) #x축범위 
plt.yticks([3,6,9,12]) #y축범위
plt.show() #이상한 문자 없앰
```
