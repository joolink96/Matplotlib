# 14. 원 그래프(심화)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic' #Windows
matplotlib.rcParams['font.size']=15 #글자크기
matplotlib.rcParams['axes.unicode_minus']=False #한글 폰트를 사용시, 마이너스 글자 깨지지 않도록함
```

```python
values=[30,25,20,13,10,2]
labels=['Python','Java','Javascript','C#','C/C++','ETC']
colors=['b','g','r','c','m','y']
explode=[0.05]*6

plt.pie(values,labels=labels,autopct='%.1f%%',startangle=90,counterclock=False,colors=colors,explode=explode) #autopct는 합이 100일때, 퍼센트로 데이터를 나타냄.
                                                                                 #90도기준으로 시계방향순으로 데이터 출력


plt.show()
```

```python
wedgeprops={'width':0.2} #wedgeprops는 도넛의 폭을 정의
plt.pie(values,labels=labels,autopct='%.1f%%',startangle=90,counterclock=False,colors=colors,explode=explode,wedgeprops=wedgeprops)
plt.show() 
```

```python
wedgeprops={'width':0.4} 
plt.pie(values,labels=labels,autopct='%.1f%%',startangle=90,counterclock=False,colors=colors,explode=explode,wedgeprops=wedgeprops)
plt.show() 
```

```python
wedgeprops={'width':0.4}  
plt.pie(values,labels=labels,autopct='%.1f%%',startangle=90,counterclock=False,colors=colors,wedgeprops=wedgeprops)
plt.show() 
```

```python
wedgeprops={'width':0.4,'edgecolor':'w','linewidth':5}  #edgecolor는 하얀색, edge의 두께는 5
plt.pie(values,labels=labels,autopct='%.1f%%',startangle=90,counterclock=False,colors=colors,wedgeprops=wedgeprops)
plt.show() 
```

```python
def custom_autopct(pct): #숫자 안겹치게 하기
    return ('%.1f%%' %pct) if pct>=10 else ''

plt.pie(values,labels=labels,autopct=custom_autopct,startangle=90,counterclock=False,colors=colors,wedgeprops=wedgeprops)
plt.show() 
```
