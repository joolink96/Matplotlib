# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic' #Windows
matplotlib.rcParams['font.size']=15 #글자크기
matplotlib.rcParams['axes.unicode_minus']=False #한글 폰트를 사용시, 마이너스 글자 깨지지 않도록함
```

```python
x=[1,2,3]
y=[2,4,8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,linewidth=10)
```

```python
plt.plot(x,y,linewidth=5)
```

## 마커

```python
plt.plot(x,y,marker='o')
```

```python
plt.plot(x,y,marker='o',linestyle='None') #선을 없애버림
```

```python
plt.plot(x,y,marker='v') 
```

```python
plt.plot(x,y,marker='v',markersize=30) 
```

```python
plt.plot(x,y,marker='X',markersize=10) 
```

```python
plt.plot(x,y,marker='v',markersize=10,markeredgecolor='red') 
```

```python
plt.plot(x,y,marker='o',markersize=20,markeredgecolor='red') 
```

```python
plt.plot(x,y,marker='o',markersize=20,markeredgecolor='red',markerfacecolor='yellow') 
```

## 선 스타일

```python
plt.plot(x,y,linestyle=':')
```

```python
plt.plot(x,y,linestyle='--')
```

```python
plt.plot(x,y,linestyle='-.')
```

## 색깔

```python
plt.plot(x,y,color='pink')
```

```python
plt.plot(x,y,color='#ff0000') #빨강색
```

```python
plt.plot(x,y,color='b') #파랑색
```

```python
plt.plot(x,y,color='g') #초록색
```

## 포맷

```python
plt.plot(x,y,'ro--') # color,marker,linestyle
```

```python
plt.plot(x,y,'bv:') # color,marker,linestyle
```

```python
plt.plot(x,y,'go') #linestyle이 없어서 선은 사라짐.
```

## 축약어

```python
plt.plot(x,y,marker='o',mfc='red',ms=20,mec='blue',ls=':') 
#mfc=markerfacecolor
#ms=markersize
#mec=markeredgecolor
#ls=linestyle
```

## 투명도

```python
plt.plot(x,y,marker='o',mfc='red',ms=20,alpha=0.3) #alpha : 투명도(0~1) 
```

## 그래프 크기

```python
plt.figure(figsize=(10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,5),dpi=200) #dpi-확대 
plt.plot(x,y)
```

## 배경색

```python
plt.figure(facecolor='yellow')  
plt.plot(x,y)
```
